```python
import numpy as np

mu, sigma =(4,2)
s1 = np.random.normal(mu, sigma, 1000)
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s1, 30, density=True)
plt.show()
```

```python
import numpy as np

mu, sigma =(3,1)
s2 = np.random.normal(mu, sigma, 1000)
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s2, 30, density=True)
plt.show()
```
